```python
import random
import conf
import importlib
import numpy as np
import mpl_interface
import matplotlib.pyplot as plt
from matplotlib.animation import ArtistAnimation
from IPython import display

importlib.reload(conf)
importlib.reload(mpl_interface)
from conf import *
from mpl_interface import *
```

```python
p = {"beta": 10, "J": 1, "L": 10}
c = Conf(**p)
```

```python
fig, ax = plt.subplots()
ax.set_xlim(0, c.L)
ax.set_xlabel("$x$")
ax.set_ylabel(r"$\tau$")
frames = []
num_frames = 100
for i in range(num_frames):
    c.conf_update()
    frame = ax.spinplot(c)
    frame.append(ax.kinkplot(c))
    frames.append(frame)
```

```python
ani = ArtistAnimation(fig, frames, interval=700)
html = display.HTML(ani.to_jshtml())
display.display(html)
plt.close()
```
